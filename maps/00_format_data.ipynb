# Method 1:
requires the quantification table (data) and a previously defined csv table with two columns, 'name' and 'label' (cell_type_id).  The latter is a table that assigns the strings from the column 'name' a numerical value in the 'label' column.  

```python
data.sample(5)
```

```python
cell_type_id
```

```python
data['cell_label']=np.nan
for cell_type,label in zip(cell_type_id['name'],cell_type_id['label']):
    data.loc[data['cell_type']==cell_type,'cell_label']=label
data['cell_label']=data['cell_label'].astype(int)
data.drop(columns=['Cell Id','cell_type'],inplace=True)
data.rename(columns={'Nucleus 3D Volume':'cellSize'},inplace=True)
data.sample(5)
```

```python
check_nan = data['cell_label'].isnull().values.any()
check_nan
```

```python
data.to_csv(outdir / 'maps_formatted_table.csv',index=False)
```

# Method 2:
Requires only the quantification table (data).  It will produce the cell_type_id table mentioned above. 

```python
data=pd.read_csv(Path('./inputs/examples/quantification_table.csv'))
data['cell_type'] = data['cell_type'].astype(str)
data['cell_label'] = data['cell_type'].astype('category').cat.codes
cell_label_dict = dict(enumerate(data['cell_type'].astype('category').cat.categories))
data.drop(columns=['Cell Id','cell_type'],inplace=True)
data.rename(columns={'Nucleus 3D Volume':'cellSize'},inplace=True)
data.sample(5)
```
